```python
#styling the graphs
m=[0,2,4,6,8]
plt.figure(figsize=(12,5),dpi=(100))
plt.plot(m)
plt.plot(m,color='green',linewidth=2,marker='^',linestyle='--',markersize=10,markeredgecolor='blue',markerfacecolor='orange')
plt.show()


#shorthand notation
#fmt='[color][marker][line]'
plt.plot(m, 'b.-')
plt.show()
```

```python
#saving a graph,title,xlabel,ylabel
import matplotlib.pyplot as plt
#temparature in a week
x=[1,2,3,4,5,6,7]
y=[36,37,38,40,39,37,38]
plt.figure(figsize=(12,5),dpi=(100))
plt.plot(x,y)
plt.plot(x,y,color='blue',linewidth=2,marker='D',linestyle='-',markersize=10,markeredgecolor='red',markerfacecolor='green')


plt.title("Temperature in a week")
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.savefig("t3.png",dpi=1000)
plt.show()
```

```python
x1=[1,16,30,42,55,68,77,88]
y1=[10,18,43,56,78,82,89,94]
x2=[2,27,47,52,57,67,72,88]
y2=[2,37,42,56,66,74,88,95]
plt.figure()
plt.plot(x1,y1,'rD-.')
plt.plot(x2,y2,'gs-.')
plt.legend(['tv','smartphone'],loc='lower right')
plt.title("Advertisement effects on sales")
plt.xlabel("medium")
plt.ylabel("sales")
plt.show()
```

```python
x=np.arange(0,4.5,0.5)
print(x)

plt.plot(x[:6],x[:6]**2,'r^-')
plt.plot(x[5:],x[5:]**2,'g^--')
plt.title('Our First Graph')
plt.xlabel('X Axis',fontdict={'name':'Magneto',
                             'size':30,
                             'color':'blue',
                             'weight':'bold',
                            })
plt.ylabel('Y Axis',fontdict={'name':'Magneto',
                             'size':30,
                             'color':'cyan',
                             'weight':'bold',
                            })

plt.grid(color="b",linestyle=":",linewidth=1)
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,4,8,12,16])
plt.show()
```

```python
#Scatter Plotting
x=np.array([2,4,1,5,3,5,2,7,6,4])
y=np.array([2,6,3,4,1,7,3,2,6,3])
x1=np.array([3,4,1,6,3,6,2,7,3,7])
y1=np.array([2,6,3,4,1,7,3,5,2,5])
plt.xlabel("age",fontdict={'family':'Algerian','size':20,'color':'blue'})
plt.ylabel("speed",fontdict={'family':'arial','size':20,'color':'blue'})
plt.title('Scatter plotting',fontdict={'family':'arial','size':20,'color':'blue'})
c=["yellow","red","green","black","blue","yellow","red","cyan","cyan","red"]
plt.scatter(x,y,color=c)
plt.scatter(x1,y1)
plt.legend(["min,max"],loc="upper left")
plt.show()
```

```python
plt.figure()
langs=["c","java","c++","python","php"]
students=[23,43,26,75,50]
x=["orange","red","pink","cyan","yellow"]
plt.bar(langs,students,color=x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([20,54,88,20])

x1=np.array(["E","F","G","H"])
y1=np.array([30,45,22,55])

plt.xlabel("class",fontdict={"family":"serif","color":"green","size":20})
plt.ylabel("counts",fontdict={"family":"serif","color":"green","size":20})
plt.title("Bar Graph",fontdict={"family":"serif","color":"green","size":20})
plt.bar(x,y,color="r",width=0.5)

plt.bar(x1,y1)    #multiple bar graph
plt.legend(["new","old"],loc="center")
plt.show()
```

```python
a=np.array([22,34,64,12,76,89,64,56,22,23,43,76,98,75,32,43,45,47,26,48,88,90,87,76,90,60,68,69,70,23])
plt.hist(a,bins=[0,25,50,75,100],color="r")
plt.title("Histogram of Result")
plt.xticks([0,25,50,75,100])
plt.xlabel('marks')
plt.ylabel('no of students')
plt.show()
```

```python
a=np.array([200,100,150,50])
lb=["cars","bikes","auto","bus"]
clr=["red","cyan","green","blue"]
plt.pie(a,labels=lb,colors=clr)
plt.legend(lb)
plt.savefig("img.jpeg",dpi=200)
plt.show()
```
